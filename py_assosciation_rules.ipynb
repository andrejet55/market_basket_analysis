```python
# Handling Missing Values
```

```python
df['days_since_prior_order'] = df['days_since_prior_order'].fillna(0) # Assuming that this is their first order
```

```python
df['aisle_category'] = df['aisle_category'].fillna(df['aisle_category'].mode()[0]) # imputing with most frequent aisle category
```

```python
df['department'] = df['department'].fillna(df['department'].mode()[0]) # imputing with most frequent department category
```

```python
df.nunique()
```

```python
df = df.drop(columns=['eval_set','department_id','user_id','product_id'])
```

```python
df['order_hour_of_day'] = pd.to_numeric(df['order_hour_of_day'], errors='coerce')
```

```python
numerical_values = df.select_dtypes(include=['int64','float64'])
numerical_values
```

```python
categorical_values = df.select_dtypes(include=['object'])
categorical_values
```

```python
categorical_values['department'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
# label encoding aisle_category and department
le = LabelEncoder()
categorical_values['aisle_category'] = le.fit_transform(categorical_values['aisle_category'])
categorical_values['department'] = le.fit_transform(categorical_values['department'])
```

```python
aisle_enc = pd.get_dummies(categorical_values['aisle'], prefix='aisle', drop_first=True)
```

```python
# product_name has too many values , so frequency encoding suits best
categorical_values['product_name_freq_enc'] = categorical_values['product_name'].map(categorical_values['product_name'].value_counts())

# we are dropping the original product_name
categorical_values = categorical_values.drop(columns=['product_name'])
```

```python
# droping the aisle as we have aisle_enc
categorical_values = categorical_values.drop(columns=['aisle'])

# here we are concatenating all the encoded columns
categorical_values_enc = pd.concat([categorical_values, aisle_enc], axis=1)
categorical_values_enc
```

```python
# checking if the size before and after encoding is same
categorical_values_enc.shape
```

```python
# convertign the df into sparseformat
categorical_values_enc_sparsed = categorical_values_enc.astype(pd.SparseDtype("int", fill_value=0))

# checking teh results
categorical_values_enc_sparsed.head()
```

```python
categorical_values_enc_sparsed.shape
```

```python
# aliginign the index
categorical_values_enc_sparsed.reset_index(drop=True, inplace=True)
numerical_values.reset_index(drop=True, inplace=True)
```

```python
# numerical and catergorical values concatinantion
cat_num_merged = pd.concat([categorical_values_enc_sparsed, numerical_values], axis=1)

print(cat_num_merged.head())
```

```python
import dask.dataframe as dd
import pandas as pd
import numpy as np
#using dask because dataset is too large
cat_num_merged_dask = dd.from_pandas(cat_num_merged, npartitions=10)

# selecting just teh numerical columns
numericals = cat_num_merged_dask.select_dtypes(include=['int64', 'float64'])

# removing these two cloumn as they are id's and they are not necessaryliy outliers
numericals = numericals.drop(columns=['order_id', 'product_id'], errors='ignore')

# using map_partitions for memory utilization
numericals = numericals.map_partitions(pd.DataFrame.astype, dtype='float32')

# IQR calculations
Q1 = numericals.quantile(0.25, axis=0)
Q3 = numericals.quantile(0.75, axis=0)
IQR = Q3 - Q1

# defining lower ans upper bounds for IQR
lower_bounds = Q1 - 1.5 * IQR
upper_bounds = Q3 + 1.5 * IQR

# converting the bounds into pandas series
lower_bounds = lower_bounds.compute()
upper_bounds = upper_bounds.compute()

# identifying outliers
outliers_below = numericals < lower_bounds.values
outliers_above = numericals > upper_bounds.values

# checking the conditions to classify as outliers
outliers_found = outliers_below | outliers_above

# finding the rows that have outliers
outliers_df = cat_num_merged_dask[outliers_found.any(axis=1)]

# computing
outliers_df = outliers_df.compute()

outliers_df.head()
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
cat_num_merged['product_name_freq_enc'] = cat_num_merged['product_name_freq_enc'].sparse.to_dense()

basket = cat_num_merged.groupby(['order_id', 'product_name_freq_enc'])['product_name_freq_enc'].count().unstack().reset_index().fillna(0)
basket.set_index('order_id', inplace=True)
# Convert quantities to 1 and 0
basket = basket.applymap(lambda x: 1 if x > 0 else 0)

# Generate frequent itemsets using Apriori
frequent_itemsets = apriori(basket, min_support=0.01, use_colnames=True)

# Generate association rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

# Displaying the results
print("Frequent Itemsets:")
print(frequent_itemsets)

print("\nAssociation Rules:")
print(rules)
```
